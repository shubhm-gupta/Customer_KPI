```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn import tree
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier

def calcMeasures(testY,predicted):
    tn, fp, fn, tp = confusion_matrix(testY, predicted).ravel()
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1 = 2*(precision*recall)/(precision+recall)
    print(precision)
    print(recall)
    print(f1)
    print((tp+tn)/(tp+tn+fp+fn))
    
def randomForestClassifier(train,test,trainY,testY):
    clf = RandomForestClassifier(n_estimators=100, max_depth=2,
                             random_state=0)
    clf = clf.fit(train, trainY)
    predicted = clf.predict(train)
    print('Train')
    calcMeasures(trainY,predicted)
    print('Test')
    predicted = clf.predict(test)
    calcMeasures(testY,predicted)
    return clf

def adaBoost(train,test,trainY,testY):
    clf = AdaBoostClassifier(n_estimators=100)
    clf = clf.fit(train, trainY)
    predicted = clf.predict(train)
    print('Train')
    calcMeasures(trainY,predicted)
    print('Test')
    predicted = clf.predict(test)
    calcMeasures(testY,predicted)
    return clf


def svmClassifier(train,test,trainY,testY):
    clf = SVC(gamma='scale')
    clf = clf.fit(train, test)
    predicted = clf.predict(train)
    print('Train')
    calcMeasures(trainY,predicted)
    print('Test')
    predicted = clf.predict(test)
    calcMeasures(testY,predicted)
    return clf

    
def naiveBayes(train,test,trainY,testY):
    gnb = GaussianNB()
    gnb_model = gnb.fit(train, trainY)
    predicted = gnb_model.predict(train)
    print('Train')
    calcMeasures(trainY,predicted)
    print('Test')
    predicted = gnb_model.predict(test)
    calcMeasures(testY,predicted)
    return gnb

def decisionTree(train,test,trainY,testY):
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(train, trainY)
    predicted = clf.predict(train)
    print('Train')
    calcMeasures(trainY,predicted)
    print('Test')
    predicted = clf.predict(test)
    calcMeasures(testY,predicted)
    return clf
```
